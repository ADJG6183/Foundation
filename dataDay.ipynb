# Day 2 - Feature Engineering

```python
df['social_media_ratio'] = df['social_media_hours'] / df['daily_screen_time_hours']
df['exercise_sleep_ratio'] = df['exercise_minutes'] / df['sleep_hours']
```

```python
df.columns
```

```python
print("Before creating features:")
print(df.columns)

df['social_media_ratio'] = df['social_media_hours'] / df['daily_screen_time_hours']
df['exercise_sleep_ratio'] = df['exercise_minutes'] / df['sleep_hours']

print("\nAfter creating features:")
print(df.columns)

print("\nCheck feature values:")
print(df[['exercise_sleep_ratio', 'social_media_ratio', 'productivity_score']].head())

print("\nCorrelation:")
print(df[['exercise_sleep_ratio', 'social_media_ratio', 'productivity_score']].corr())
```

```python
df[['exercise_sleep_ratio','social_media_ratio']].isnull().sum()
```

```python
df['balance'] = df['study_work_hours'] - df['social_media_hours']
df['total_activity'] = (df['study_work_hours'] + df['exercise_minutes']) / 60

df[['total_activity', 'balance']].head()

```

```python
# Correlation between total_activity, balance, and productivity_score
df[['total_activity', 'balance', 'productivity_score']].corr()
```
